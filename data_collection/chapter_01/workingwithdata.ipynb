Cleaning and Structuring Data

```python
# ------------------------------------------Assignment 1---------------------------------------------------

import json
```

```python
def load_data(fileName):
    with open(fileName, 'r') as f:
        d1 = json.load(f)

    return d1
```

```python
data = load_data("data.json")
print(data)
print(type(data))
```

```python
# problems in data:
    #1. data is inconsistent -> (ratings)
    #2. missing data -> (age)
    #3. duplicate data -> (alice appear 2 times)
```

```python
# 1. Problem 1 

def clean_data(data):
    text_to_num = {"one": '1', "two": '2', "three": '3', "four": '4', "five": '5'}
    cleaned_data = []
    unique_user = set()

    for user in data:
        raw_rating = user["rating"].strip().lower() # remove trailing spaces

        if raw_rating in text_to_num:
            raw_rating = text_to_num[raw_rating]

        user["rating"] = raw_rating

# 2. Problem 2
        raw_age = user.get("age") # we are not using user["age"] here bcoz if age is not present in any user data, it will throw error, but if we use it, it will either return data or none
        if(raw_age == None):
            user["age"] = None

# 3. Problem 3
        if(user["name"].strip() in unique_user):
            continue

        cleaned_data.append(user)
        unique_user.add(user["name"].strip())

    return cleaned_data
```

```python
data = clean_data(data)
```

Get Meaningfull Insights from data

```python
# Insights:
    # 1. get avg ratings
    # 2. 
```

```python
def get_insights(data):

    total_rating = 0
    poor_rating = 0

    for user in data:
        total_rating += float(user["rating"])

        if(float(user["rating"]) < 3.0):
            poor_rating+=1
    
    avg_rating = total_rating/(len(data))
    perc_poor = poor_rating/(len(data))

    print(f"Avg rating: {avg_rating}")
    print(f"% of user with poor rating: {perc_poor*100}%")
```

```python
get_insights(data)
```

Build a recommendation feature (rule based)

```python
def get_recommendations(data):
    recommendations = []

    for user in data:
        curr_recom = {}
        curr_recom["name"] = user["name"]

        if(float(user["rating"]) >= 4.0):
            curr_recom["brand"] = "Apple"
        else:
            curr_recom["brand"] = "Samsung"

        recommendations.append(curr_recom)
    
    return recommendations
```

```python
get_recommendations(data)
```
